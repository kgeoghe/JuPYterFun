Let's plot two sinusoids added together: $y = \cos{(f0*t + a0)}+ \cos{(f1*t + a1)}$.  That's always fun at a party.

```python
import bokeh.models as bm
import bokeh.plotting as bp
import bokeh.io as bio
import ipywidgets as ipw
import numpy as np
```

```python
# we have to tell bokeh where it is.  Otherwise bokeh will put plots in other browser tabs 😒
bio.output_notebook()
```

```python
# this is the function we're wanting to visualize
def add_cosines(t, f0, a0, f1, a1):
    """What even happens when you add sinusoids?"""
    y = np.cos(f0*t + a0) + np.cos(f1*t + a1)
    return y
```

```python
# Evaluate the sinusoids at these points
t=np.linspace(-np.pi*2, np.pi*2, 1000)

fig = bp.figure(x_range=[-np.pi*2, np.pi*2], y_range=[-2, 2])
line = fig.line([], [])
data_source = line.data_source

def callback(f0, a0, f1, a1):
    y = add_cosines(t, f0, a0, f1, a1)
    data_source.data = {'x': t, 'y': y}
    bio.push_notebook(handle)

# this time I'm passing **widget instances** instead of just values.
# This gives us more control over how things are presented, and the values
# the controls can take
interacter = ipw.interactive(
    callback,
    f0=ipw.BoundedFloatText(description='f0', value=1, min=-20, max=20, step=0.05),
    a0=ipw.BoundedFloatText(description='a0', vlaue=0, min=-np.pi, max=np.pi, step=0.05),
    f1=ipw.BoundedFloatText(description='f1', value=0.5, min=-20, max=20, step=0.05),
    a1=ipw.BoundedFloatText(description='a1', vlaue=0, min=-np.pi, max=np.pi, step=0.05),
)
# there are other widgets we cold use

# gotta grab a handle so we can update the plot later
handle = bio.show(fig, notebook_handle=True)
interacter
```
